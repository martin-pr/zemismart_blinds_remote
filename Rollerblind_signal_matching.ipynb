```python
import matplotlib.pyplot as plt

# Plot the signal from the
def plotSignal(signal: list[int], limit = -1):
  t = [0]
  vals = [0]
  for i in signal:
    i = int(i)
    current = t[-1]

    if current > limit and limit > 0:
      vals.pop()
      t.pop()
      break

    if i < 0:
      t.append(current + 0.01)
      vals.append(0)
      t.append(current - i)
      vals.append(0)
    else:
      t.append(current + 0.01)
      vals.append(1)
      t.append(current + i)
      vals.append(1)

  fig, ax = plt.subplots(figsize=(30, 3), dpi=400)

  ax.plot(t, vals, linewidth=0.5)
  ax.set_xlim(0, t[-1])
  plt.show()
```
